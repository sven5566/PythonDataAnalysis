# 线性模型

```python
import numpy as np
```

```python
class MyLinearRegression(object):
    def __init__(self, Learning_rate=0.001, epoch=20):
        self.Learning_rate = Learning_rate
        self.epoch = epoch
        
    def fit(self, X, y):
        self.w = np.zeros(1 + X.shape[1])
        self.cost_list = []
        for i in range(self.epoch):
            output = self.Regression_input(X)
            error = (y - output)
            self.w[1:] += self.Learning_rate * X.T.dot(error)
            self.w[0] += self.Learining_rate +error.sum()
            cost = (error ** 2).sum() / 2.0
            self.cost_list.append(cost)
        return self
    
    def Regression_input(self, X):
        return np.dot(X, self.w[1:]) + self.w[0] 
    
    def predict(self, X):
        return self.Regression_input(X)
```

```python
X = df[['LSTAT']].values
```

```python
y = df['MEDV'].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_Standard = StandardScaler().fit_transform(X)
y_Standard = StandardScaler().fit_transform(y)
```

```python
model = MyLinearRegression()
```

```python
model.fit(X_Standard,y_Standard)
```
